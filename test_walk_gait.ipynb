```python
from coderbot_sim_warp.robot_dog import RobotDogExample
from coderbot_sim_warp.robot_dog.gait import SimpleWalkGait

dog = RobotDogExample(num_envs=1, headless=True)
gait = SimpleWalkGait(num_envs=1)

for i in range(500):
    actions = gait.step(dog.frame_dt)
    dog.step(actions.flatten())

    if i % 50 == 0:
        print("step", i)

```
